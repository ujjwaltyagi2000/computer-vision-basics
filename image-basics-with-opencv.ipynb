* The above image has a blue tint over it.
* This is because matplotlib and opencv have different orders of the R G B channels.
* matplotlib --> R-G-B
* OpenCV --> B-G-R
**When imread reads a file, it reads it as BGR, but when plotting, matplotlib plots it as RGB. So the Red and Blue channels are getting swapped**

We therefore need to transform this to go from BGR to RGB, the way matplotlib expects it.

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
#To resize:
new_img = cv2.resize(fix_img,(500,350))
#Takes a tuple as argument containing dimensions (width, height)
```

```python
plt.imshow(new_img)
```

#### **Important Note: In NumPy, resolution is taken as (height, width). OpenCV follows the opposite convention where it takes resolution as (width, height)**

**Resize By ratio:**

```python
w_ratio = 0.8
h_ratio=0.4
```

```python
new_img=cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
#The syntax for this is weird
```

```python
plt.imshow(new_img)
```

### Flip image horizontally and vertically:

```python
flip_img=cv2.flip(fix_img,0) #0 --> Horizontal flip
plt.imshow(flip_img)
```

```python
vflip_img=cv2.flip(fix_img,1) #1 --> Vertical fLip
plt.imshow(vflip_img)
```

## Saving Images:
