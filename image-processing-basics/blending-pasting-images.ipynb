# BLENDING AND PASTING IMAGES

Often we work with multiple images. OpenCV has many programmatic methods of blending images together and pasting images on top of each other.

Blending images is done through the __addWeighted__ function that uses both images and combines them.
To blend images we use a simple formula:
> new_pixel = $alpha$ x pixel_1 + $beta$ x pixel_2 + $gamma$    

### __Syntax to do this:__

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('../DATA/dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('../DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

The size of both images is different

```python
shape1=img1.shape
shape1
```

```python
shape2=img2.shape
shape2
```

**To blend two images, both images must be of same size**

Let's resize both images to 1200*1200

```python
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended_img = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.1,gamma=10)
#try using different values of alpha beta and gamma
```

```python
plt.imshow(blended_img)
```

#### Overlay small image on top of a larger image (No Blending)

* Will use NumPy reassignment

```python
#reloading images again
large_img = cv2.imread('../DATA/dog_backpack.jpg')
large_img = cv2.cvtColor(large_img, cv2.COLOR_BGR2RGB)
small_img = cv2.imread('../DATA/watermark_no_copy.png')
small_img = cv2.cvtColor(small_img, cv2.COLOR_BGR2RGB)
```

```python
small_img= cv2.resize(small_img,(600,600))
```

```python
plt.imshow(small_img)
```

```python
plt.imshow(large_img)
```

```python
#let's define offset coordinates: from where the overlaying will start
x_offset=0
y_offset=0
#we also need to define end points: offset +  length of small image
x_end=x_offset+small_img.shape[1]
y_end=y_offset+small_img.shape[0]
```

For NumPy:
* columns --> X-values
* rows --> Y-values

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
#It's like replacing elements of a matrix with values of another matrix
```

```python
plt.imshow(large_img)
```
